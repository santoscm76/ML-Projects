```
df_append = df_1.copy().append([df_2.copy(), df_3.copy(), df_4.copy(), df_5.copy(), df_6.copy(), df_7.copy(), df_8.copy(), df_9.copy(), df_10.copy(), df_11.copy(), df_12.copy(), df_13.copy(), df_14.copy(), df_15.copy(), df_16.copy(), df_17.copy(), df_18.copy(), df_19.copy(), df_20.copy()])
```

```
df_no_duplicate = df_append.drop_duplicates(keep='first')
```

```
df_3_columns = df_no_duplicate[["datetime_value", "grid_id", "C01"]]
test = df_3_columns.pivot_table(index='datetime_value', columns='grid_id', values='C01', fill_value=0)
test_2 = pd.DataFrame(test.to_records())
test_2['row_number'] = test_2.reset_index().index
```

```
df_input = test_2[test_2['row_number'] < 286]
test_3 = test_2[['cmlgrid_990', 'row_number']]
df_target = test_3[test_3['row_number'] > 0]

df_target['new_row_number']  = df_target.copy()['row_number'] -1

```

```
df_input.reset_index(drop=True)
df_target.reset_index(drop=True)
df_concat = pd.merge(left=df_input, right=df_target, left_on='row_number', right_on='new_row_number')
df_concat.rename(columns={'cmlgrid_990_y': 'target', 'row_number_x': 'row_number'}, inplace=True)
```

```
test_4 = df_concat.copy().drop(['datetime_value', 'row_number_y', 'new_row_number'], axis=1)
df_concat = test_4
```

```
train_X = df_concat.copy().drop(columns=['target'])
train_y = df_concat.copy()[['target']]
```

```
from keras.models import Sequential
from keras.layers import Dense

#create model
model = Sequential()

#get number of columns in training data
n_cols = train_X.shape[1]

#add model layers
model.add(Dense(4155, activation='relu', input_shape=(n_cols,))) #4155
model.add(Dense(4155, activation='relu'))
model.add(Dense(4155, activation='relu'))
model.add(Dense(4155, activation='relu'))
model.add(Dense(1))

#compile model using mse as a measure of model performance
model.compile(optimizer='adam', loss='mean_squared_error')
```

```
from keras.callbacks import EarlyStopping
#set early stopping monitor so the model stops training when it won't improve anymore
early_stopping_monitor = EarlyStopping(patience=3)
#train model
model.fit(train_X, train_y, validation_split=0.2, epochs=30, callbacks=[early_stopping_monitor])
```

```
# Auto Keras
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_X, train_y, test_size=0.3)

import autokeras as ak
search = ak.StructuredDataClassifier(max_trials=150)
search.fit(x=X_train, y=y_train, verbose=1/0)
```
