```
import time
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

tf.__version__

#url = 'https://raw.githubusercontent.com/sniafas/ML-Projects/master/Regression/data/birth_rate.csv'
#data = pd.read_csv(url)

from google.colab import files
uploaded = files.upload()
import io
data = pd.read_csv(io.BytesIO(uploaded['data_train.csv']))

data.head()
```

```
#data_X = np.array(data['Birth rate'])
#data_Y = np.array(data['Life expectancy'])

data_X = np.array(data['X_Value'])
data_Y = np.array(data['Y_Value'])
```

```

# Weight and Bias

w = tf.Variable(tf.zeros(1), name="weight")
b = tf.Variable(tf.zeros(1), name="bias")

def predict(x):
    return x * w + b

def mse(y_true, y_pred):
    return tf.losses.mean_squared_error(y_true,y_pred)

# SGD Optimizer
optimizer = tf.optimizers.SGD(learning_rate=0.01)
```

```
display_step = 500
learning_rate = 0.001
epochs = 5000

history = []
start = time.time()

# Start training
for epoch in range(1, epochs + 1):
    
    # Begin GradientTape and optimise
    with tf.GradientTape() as g:
        pred = predict(data_X)
        loss = mse(data_Y, pred)

        #print("Epoch ", epoch, ". pred =", pred, ". loss = ", loss)

    # Compute dw, db
    gradients = g.gradient(loss, [w,b])
    
    # Update w and b
    optimizer.apply_gradients(zip(gradients, [w,b]))
    
    # Monitor training / display steps
    if epoch % display_step == 0:
        pred = predict(data_X)
        loss = mse(pred, data_Y)
        history.append(np.array(w * data_X + b))
        print("Epoch: %i, Loss: %f, w: %f, b: %f" % (epoch, loss, w.numpy(), b.numpy()))
print('Train Time: %f seconds' %(time.time() - start))
```
